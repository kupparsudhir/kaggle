```python
def missing_value(df):
    col = df.columns
    for i in col:
        if df[i].isnull().sum()>0:
            df[i].fillna(df[i].mode()[0],inplace=True)
```

```python
missing_value(train)
missing_value(test)
```

# One Hot Encoding

```python
def OHE(df):
    cat_col = [col for col in df.columns if '_cat' in col]
    bin_col = [col for col in df.columns if '_bin' in col]
    
    c2,c3 = [],{}
    print('Binary feature',len(bin_col))
    for c in  bin_col:
        if df[c].nunique()>2 :
            c2.append(c)
            c3[c] = 'ohe_'+c
            print(c3,c2)
    print('Categorical feature',len(cat_col))
    for c in cat_col:
        if df[c].nunique()>2 :
            c2.append(c)
            c3[c] = 'ohe_'+c
    
    df = pd.get_dummies(df, prefix=c3, columns=c2,drop_first=True)

    df = df.drop(bin_col,axis=1)
    #df = df.drop(cat_col,axis=1)
    print(df.shape)
    return df
```

```python
train1 = OHE(train)
test1 = OHE(test)
```

```python
#split data set
X = train1.drop(['target','id'],axis=1)
y = train1['target']
x_test = test1.drop('id',axis=1)
xtr,xvl,ytr,yvl = train_test_split(X,y,test_size=0.3,random_state=192)
```

```python
#Modelling
lr = LogisticRegression(max_iter=1000,verbose = 1,random_state=seed)
lr.fit(xtr,ytr)

y_lr = lr.predict(xvl)
```

```python
confusion_matrix(yvl,y_lr)
```

```python
roc_auc_score(yvl,lr.predict_proba(xvl)[:,1])
```
